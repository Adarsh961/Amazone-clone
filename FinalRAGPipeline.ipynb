## Modify retrieval

```python
def retrieve(query, index, chunks, embed_model, top_k=3):
    query_embedding = embed_model.encode([query], normalize_embeddings=True)
    D, I = index.search(np.array(query_embedding), top_k)
    print(f"ğŸ” Top Chunks for '{query}': {[chunks[i][:60] for i in I[0]]}")  # Debug
    results = [chunks[i] for i in I[0]]
    distances = D[0]  # Get the distances for the top-k results
    return results, distances
```

## Implement relevance filtering

### Subtask:
Create a new function to filter the retrieved chunks based on a similarity threshold.


**Reasoning**:
Define a function to filter chunks based on their similarity distance to a query.



```python
def filter_chunks_by_similarity(chunks, distances, threshold):
    """
    Filters retrieved chunks based on a similarity distance threshold.

    Args:
        chunks: A list of text chunks.
        distances: A list of corresponding similarity distances (lower is more similar).
        threshold: The maximum distance allowed for a chunk to be kept.

    Returns:
        A list of filtered chunks.
    """
    filtered_chunks = [chunks[i] for i in range(len(chunks)) if distances[i] <= threshold]
    return filtered_chunks
```

## Update qa generation

### Subtask:
Modify the `generate_answer` function to use the filtered chunks.


**Reasoning**:
Modify the `generate_answer` function to accept and use the filtered chunks.



```python
def generate_answer(filtered_context_chunks, question):
    """
    Generates an answer to a question based on provided context chunks.

    Args:
        filtered_context_chunks: A list of text chunks filtered by relevance.
        question: The question to answer.

    Returns:
        The generated answer string.
    """
    context = " ".join(filtered_context_chunks)
    input_text = f"question: {question} context: {context}"
    input_ids = qa_tokenizer.encode(input_text, return_tensors='pt', truncation=True, max_length=1024).to(device)

    output_ids = qa_model.generate(input_ids, max_length=256, num_beams=4, early_stopping=True)
    return qa_tokenizer.decode(output_ids[0], skip_special_tokens=True)
```

## Integrate into gradio interface

### Subtask:
Update the Gradio interface to use the modified functions.


**Reasoning**:
Update the `process_pdf` function to use the modified `retrieve` function, call the new filtering function, and pass the filtered chunks to `generate_answer`.



```python
import gradio as gr

def toggle_question(choice):
    return gr.update(visible=choice == "Q&A")

def process_pdf(pdf_file, question):
    import fitz
    from nltk.tokenize import sent_tokenize
    import numpy as np

    # Step 1: Read up to 10 pages from PDF
    doc = fitz.open(pdf_file.name)
    pages = [page.get_text() for page in doc[:10]]
    full_text = " ".join(pages)

    sentences = sent_tokenize(full_text)
    prepared_text = " ".join(sentences)

    chunks = chunk_text_fixed_token(prepared_text)
    embeddings = embed_chunks(chunks)
    index = build_faiss_index(np.array(embeddings))

    # Retrieve chunks and distances using the updated retrieve function
    retrieved_chunks, distances = retrieve(question, index, chunks, embed_model)

    # Filter chunks based on a similarity threshold (adjust threshold as needed)
    similarity_threshold = 0.7 # Example threshold, can be refined
    filtered_chunks = filter_chunks_by_similarity(retrieved_chunks, distances, similarity_threshold)

    # Generate answer using the filtered chunks
    if filtered_chunks:
        answer = generate_answer(filtered_chunks, question)
        return f"ğŸ’¬ Answer:\n{answer}"
    else:
        return "âŒ No relevant information found for your question."
```
